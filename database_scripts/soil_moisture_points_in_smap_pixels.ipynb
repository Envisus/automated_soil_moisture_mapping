```python
import psycopg2

conn = psycopg2.connect('')
cur = conn.cursor()
```

```python
query = 'SELECT ease_column_index, ease_row_index FROM smap_grid_ok;'
cur.execute(query)

smap_pixels = cur.fetchall()
```

```python
query = 'SELECT COUNT(sm.id) FROM smap_grid_ok AS smap, soil_moisture_grid_id AS sm, state_ok AS ok ' +\
        'WHERE (' +\
            'ST_Intersects(smap.geom_pixel, sm.geom_point) AND ' +\
            'ST_Intersects(ok.geom, sm.geom_point) AND ' +\
            'smap.ease_column_index = %s AND ' +\
            'smap.ease_row_index = %s' +\
        ');'   
```

```python
points = []
for smap_pixel in smap_pixels:
    cur.execute(query, smap_pixel)
    points.append(cur.fetchone()[0])
```
